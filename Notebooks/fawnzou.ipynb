```python
# SQLAlchemy
from sqlalchemy import create_engine, inspect
import pandas as pd
# Path to sqlite

# Path to sqlite
database_path = "../Data/wdi_kaggle.sqlite"

# Create an engine that can talk to the database
engine = create_engine(f"sqlite:///{database_path}")
#from sqlalchemy import inspect
for table_name in inspect(engine).get_table_names():
   print(table_name)
```

```python
conn= engine.connect()
Country_df= pd.read_sql("select * from Country",conn)
CountryNotes_df= pd.read_sql("select * from Country",conn)
Footnotes_df= pd.read_sql("select * from Country",conn)
Indicators_df= pd.read_sql("select * from Country",conn)
Series_df= pd.read_sql("select * from Country",conn)
SeriesNotes_df= pd.read_sql("select * from Country",conn)
```

```python
# Save all dataframes as csv files
Country_df.to_csv("out_put\Country.csv",index=False)
CountryNotes_df.to_csv("out_put\CountryNotes.csv",index=False)
Footnotes_df.to_csv("out_put\Footnotes.csv",index=False)
Indicators_df.to_csv("out_put\Indicators.csv",index=False)
Series_df.to_csv("out_put\Series.csv",index=False)
SeriesNotes_df.to_csv("out_put\SeriesNotes.csv",index=False)
```
